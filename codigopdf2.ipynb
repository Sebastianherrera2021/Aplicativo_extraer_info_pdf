## *Convertir pdf en txt*

```python
# Definir la carpeta que contiene los archivos PDF
carpeta_pdfs = r"C:\Users\juan2\Downloads\pruebina"  # Cambia esto a la ruta de tu carpeta

# Inicializar listas para almacenar los datos
data = []

# Función para procesar un solo PDF
def procesar_pdf(archivo):
    with pdfplumber.open(archivo) as pdf:
        texto_total = ""
        for page in pdf.pages:
            texto = page.extract_text()
            if texto:
                texto_total += texto + "\n"  # Acumular el texto total
        return texto_total.lower()  # Devolver el texto en minúsculas

##################################################################################################
# Procesar todos los archivos PDF en la carpeta
for archivo in os.listdir(carpeta_pdfs):
    if archivo.endswith('.pdf'):
        procesar_pdf(os.path.join(carpeta_pdfs, archivo))

# Previsualizar el contenido del archivo .txt
with open("prueba.txt", "r", encoding="utf-8") as preview_file:
    contenido = preview_file.read()  # Leer todo el contenido del archivo
    print("Previsualización del contenido del archivo:")
    print(contenido)

print("Conversión completada. El archivo se ha guardado como 'prueba.txt'")
```

## *Plazo de ejecucion y objeto*

```python
# Cargar el archivo
#with open('prueba.txt', encoding="utf8") as file:
    #text = file.read().lower()

# Definir las palabras/frases clave para un solo punto
palabras_con_un_punto = [' CONTRATO DE PRESTACION DE SERVICIOS No.','plazo de ejecución:', 'objeto:']

# Diccionario para almacenar el texto después de cada palabra clave
palabras_con_dos_puntos = ['valor del contrato:', 'obligaciones:']

# Inicializar una lista para almacenar los resultados
resultados = []
```

```python
# Procesar todos los archivos PDF en la carpeta
for archivo in os.listdir(carpeta_pdfs):
    if archivo.endswith('.pdf'):
        texto_pdf = procesar_pdf(os.path.join(carpeta_pdfs, archivo))  # Procesar el PDF

        # Contar cuántas veces aparece cada palabra/frase
        for palabra in palabras_con_un_punto + palabras_con_dos_puntos:
            coincidencias = re.findall(re.escape(palabra), texto_pdf, re.IGNORECASE)
            frecuencia = len(coincidencias)

            # Extraer el texto después de las palabras clave
            if palabra in palabras_con_un_punto:
                patron = re.compile(re.escape(palabra) + r'\s*([^.]+)\.', re.IGNORECASE)
                texto_extraido = patron.findall(texto_pdf)
                texto_extraido = texto_extraido[0].strip() if texto_extraido else None
            else:  # Para palabras con dos puntos
                patron = re.compile(re.escape(palabra) + r'\s*(.*?)(?=\.\.)', re.IGNORECASE | re.DOTALL)
                texto_extraido = patron.findall(texto_pdf)
                texto_extraido = texto_extraido[0].strip() if texto_extraido else None
            
            # Añadir los resultados a la lista
            resultados.append({
                'Archivo': archivo,
                'Clave': palabra,
                'Frecuencia': frecuencia,
                'Información': texto_extraido
            })
```

```python
# Crear DataFrame a partir de los resultados
df = pd.DataFrame(resultados)
print(df)
```

## *Exportar a excel*
